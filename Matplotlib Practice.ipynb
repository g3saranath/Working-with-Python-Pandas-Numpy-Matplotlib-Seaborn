```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.linspace(0,5,11)
y= x**2
x,y
```

```python
#Functional Method
plt.plot(x,y,'g-')
plt.xlabel("X Values")
plt.ylabel("Squares of X")
plt.title('X vs Square of X')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b')
plt.subplot(1,2,2)
plt.plot(x,y,'r')
```

```python
#Object Oriented Method
fig=plt.figure() #Creates a figure object
axes=fig.add_axes([0.2,0.2,1,1])
axes.plot(x,y)
axes.set_xlabel('Values of X')
axes.set_ylabel('Squares of X')
axes.set_title("Plotting using Object Oriented API Method")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.24,0.5,0.3,0.2]) #35% from left, 50% from bottom, 30% of width and 20% of height
axes1.plot(x,y)
axes1.set_title('Axes1')
axes2.plot(y,x)
axes2.set_xlim([20,30])
axes2.set_ylim([3,5])
axes2.set_title('Axes2')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
plt.tight_layout()
axes[1].plot(x,y)
axes[1].set_title("2nd")
axes[0].plot(y,x)
axes[0].set_title("1ST Plot")
```
