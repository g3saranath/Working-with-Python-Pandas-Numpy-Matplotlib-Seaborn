```python
import seaborn as sns
%matplotlib inline
import numpy as np
tips=sns.load_dataset('tips')
flight=sns.load_dataset('flights')
```

```python
flight
```

```python
tc=tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fl=flight.pivot_table(index='year',columns='month',values='passengers')
```

```python
sns.heatmap(fl,cmap='magma',linecolor='white',linewidths=1,annot=True)
```

```python
sns.lineplot(x='year',y='passengers',data=flight)
```

```python
sns.clustermap(fl,standard_scale=1)
```

```python
import seaborn as sns
%matplotlib inline
import numpy as np
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g=sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.displot)
plt.tight_layout()
```

```python
tips=sns.load_dataset('tips')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['s','^'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='time',aspect=2,size=5)
```
